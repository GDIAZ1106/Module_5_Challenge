### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_city = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_city.head()
```

```python
# 2. Get the total drivers for each city type
total_drivers_city = city_data_df.groupby(["type"]).sum()['driver_count']
total_drivers_city
```

```python
#  3. Get the total amount of fares for each city type
total_fares_city = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_city 
```

```python
#  4. Get the average fare per ride for each city type. 
average_fares_city = total_fares_city / total_rides_city
average_fares_city
```

```python
# 5. Get the average fare per driver for each city type. 
average_fares_driver = total_fares_city / total_drivers_city 
average_fares_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_ride_summary_df = pd.DataFrame({
    "Total Rides" : total_rides_city,
    "Total Drivers" : total_drivers_city,
    "Total Fares" : total_fares_city,
    "Average Fare Ride": average_fares_city,
    "Average Fare Driver": average_fares_driver,
})

pyber_ride_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_ride_summary_df.index.name = None
pyber_ride_summary_df
```

```python
#  8. Format the columns.

# Format the "Total Fares" to have the comma for a thousands separator, a decimal separator and a "$".
# Dim total_fares_city as number
pyber_ride_summary_df['Total Rides'] = pyber_ride_summary_df['Total Rides'].map("{:,.2f}".format)
pyber_ride_summary_df["Total Drivers"] = pyber_ride_summary_df["Total Drivers"].map("{:,.2f}".format)
pyber_ride_summary_df['Total Fares'] = pyber_ride_summary_df['Total Fares'].map("${:,.2f}".format)
pyber_ride_summary_df["Average Fare Ride"] = pyber_ride_summary_df["Average Fare Ride"].map("${:,.2f}".format)
pyber_ride_summary_df["Average Fare Driver"] = pyber_ride_summary_df["Average Fare Driver"].map("${:,.2f}".format)

# Display the data frame (check if we are going ok)
pyber_ride_summary_df

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

fares_city_date_df = pyber_data_df.groupby(['type', 'date']).sum()
fares_city_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_city_date_df = fares_city_date_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. ''
sum_fare_by_type_pivot = fares_city_date_df.pivot(index="date", columns="type", values="fare")
sum_fare_by_type_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04
fares_jan_apr = sum_fare_by_type_pivot.loc['2019-01-01':'2019-04-29']
fares_jan_apr.head(10)
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fares_jan_apr.index = pd.to_datetime(fares_jan_apr.index)
fares_jan_apr
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fares_jan_apr.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = fares_jan_apr.resample("W").sum()
weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

ax = weekly_fares_df.plot(figsize = (20, 5))
ax.set_title("Total Fare by City Type Jan-Apr 2019")
ax.set_xlabel('Month')
ax.set_ylabel('Fare $USD')
plt.savefig("Resources/Challenge_fig.png")
```

```python
# 9. Additional Chart for analysis. 
fares_jan_dic = sum_fare_by_type_pivot.loc['2019-01-01':'2019-12-31']
fares_jan_dic.index = pd.to_datetime(fares_jan_dic.index)
weekly_fares_annual_df = fares_jan_dic.resample("W").sum()
ax = weekly_fares_annual_df.plot(figsize = (20, 5))
ax.set_title("Total Fare by City Type Jan-Dic 2019")
ax.set_xlabel('Month')
ax.set_ylabel('Fare $USD')
plt.savefig("Resources/Challenge_fig2.png")
```
